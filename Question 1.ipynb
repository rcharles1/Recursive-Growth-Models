```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def population(n):
    if n == 0:
        return 0
    else:
        return 200 + 0.5*(1-200)/2000*200 + population(n-1)
        
    
```

```python
x = []
y = []
for n in range(0,500):
    y.append(population(n))
    x.append(n)
      
```

```python
plt.plot(x,y)
plt.title("population Vs Years")
plt.ylabel("Population")
plt.xg label('Years')
```
