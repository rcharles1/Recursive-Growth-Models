```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import sys
sys.setrecursionlimit(10**8)
def f(n):
    if n==0:
        return 0
    else:
        return(600 + 2.05*(1-600)/1000*600 * f(n-1))
```

```python
x = []
y = []
for n in range(0,50):
    y.append(f(n))
    x.append(n)
```

```python
plt.plot(x,y)
plt.xlabel('Number of Years')
plt.ylabel('Population')
plt.title("Lizards population growth")
```
